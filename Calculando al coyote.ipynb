# CALCULANDO AL COYOTE
Considere que se modela el sistema de la piedra que cae de la siguiente manera (2D):

Un cuerpo cuadrado de b=3[m] de masa m=2000[kg] que pivota sobre la esquina (punto O). Suponga que la piedra rueda sin deslizar y que hay una fricción en la rueda con µ=0.3.



Calcule:
- El ángulo al cual el cuerpo pierde contacto con la esquina (punto O) y empieza a caer 
> * Para esto, va a necesitar realizar un gráfico de la variación de la normal con respecto al ángulo de giro*
- Y ya que estamos también realice un gráfico de la variación de la velocidad angular hasta el mismo ángulo del literal anterior
- ¿Podría pasar lo que pasó en el vídeo?

Resolución:

En primer lugar vamos a realizar un diagrama de cuerpo libre, sobre el cuerpo: 



Ahora realizamos la sumatoria de fuerzas en cada eje:

Sumatoria de Fuerzas en el eje x:

$$\sum{F_x}=0$$

$$mgsen{\theta}-F_{r}=0$$

Por definición: $F_{r}=uN$

Reemplazamos:

$$mgsen{\theta}-uN=0$$

Sumatoria de Fuerzas en el eje y:

$$\sum{F_y}=0$$

$$N-mgcos{\theta}=0$$

De donde: $N=mgcos{\theta}$

Reemplazamos este valor en la sumatoria de fuerzas de x, obteniendo: 

$$mgsen{\theta}=umgcos{\theta}$$

Despejamos el valor de $\theta$ reemplazando los valores del ejercicio:

$$tan\theta=0,3$$

$$\theta=16,69°$$


Gráfico de la variación de la normal con respecto al ángulo de giro: 

```python
import numpy as np #importar libreria matematica
import plotly.express as px #herramienta de plot
import plotly.graph_objects as go
a=np.linspace(0,20,600)
N=(2000*9.8)*np.cos(np.radians(a)) #transportando al plano xy inercial. 
fig=px.line(x=a,y=N,labels={'x':'ángulo de giro [°]','y':'Normal[N]'},title="Variación de la normal con respecto al ángulo de giro")
fig.show()
```

Ahora calculamos la inercia:

$$I=\frac{1}{12}m(b^2+b^2)+mb^2$$

$$I=\frac{1}{12}2000(2(3^2))+2000(3^2)$$

$$I=7500$$

Ahora calculamos la sumatoria de torques alrededor del punto O:

$$\sum{\varGamma_ext}=I\alpha$$

$$-mg\frac{b}{2}sen(\theta)\vec{k}=I\dot{w}$$

Integramos: 

$$\int_{180}^{180-\theta}  \! -mg\frac{b}{2}sen(\theta) \, d\theta = \int_{0}^{w}  \! Iw \, dw $$

$$\frac{mgb}{I}(cos(180-\theta)-cos(180))=w^2$$

$$w^2=\frac{mgb}{I}(-cos(\theta)+1)$$

Por lo tanto, obtenemos la siguiente función:

$$w=\sqrt{\frac{mgb}{I}(-cos(\theta)+1)}$$

Gráfico de la variación de la velocidad angular hasta el ángulo de perdida de contacto

```python
import numpy as np #importar libreria matematica
import plotly.express as px #herramienta de plot
import plotly.graph_objects as go
a=np.linspace(0,20,600)
w=np.sqrt(-7.85*np.cos(np.radians(a))+7.85)
fig=px.line(x=a,y=w,labels={'x':'ángulo de giro [°]','y':'Velocidad angular [rad/s]'},title="Velocidad angular con respecto al ángulo de giro")
fig.show()
```

Y si, es posible que suceda el evento del video, solo si el ángulo de giro es mucho menor a 16,69°
